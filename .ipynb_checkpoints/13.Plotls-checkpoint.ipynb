```julia
using Plots
plot()                                      # empty Plot object
```

```julia
plot(4)                                      # initialize with 4 empty series
```

```julia
plot(rand(10))                               # 1 series... x = 1:10
```

```julia
plot(rand(10,5))                             # 5 series... x = 1:10
```

```julia
plot(rand(10), rand(10))                     # 1 series
```

```julia
plot(rand(10,5), rand(10))                   # 5 series... y is the same for all
```

```julia
plot(sin, rand(10))                          # y = sin.(x)
```

```julia
plot(rand(10), sin)                          # same... y = sin.(x)
```

```julia
plot([sin,cos], 0:0.1:)                     # 2 series, sin.(x) and cos.(x)
```

```julia
plot([sin,cos], 0, )                        # sin and cos on the range [0, ]
```

```julia
plot(1:10, Any[rand(10), sin])               # 2 series: rand(10) and map(sin,x)
```

```julia
plot!(title = "New Title", xlabel = "New xlabel", ylabel = "New ylabel")
plot!(xlims = (0, 5.5), ylims = (-2.2, 6), xticks = 0:0.5:10, yticks = [0,1,5,10])
```

```julia
# or using magic:
plot!(xaxis = ("mylabel", :log10, :flip))
xaxis!("mylabel", :log10, :flip)
```

```julia
using Plots
x = 1:10; y = rand(10); # These are the plotting data
plot(x, y)
```

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y)
```

```julia
z = rand(10)
plot!(x, z)
```

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
p = plot(x, y)
z = rand(10)
plot!(p, x, z)
```

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y, title = "Two Lines", label = ["Line 1" "Line 2"], lw = 3)
```

```julia
gr() # We will continue onward using the GR backend
plot(x, y, seriestype = :scatter, title = "My Scatter Plot")
```

```julia
scatter(x, y, title = "My Scatter Plot")
```

```julia
y = rand(10, 4)
plot(x, y, layout = (4, 1))
```

```julia
p1 = plot(x, y) # Make a line plot
p2 = scatter(x, y) # Make a scatter plot
p3 = plot(x, y, xlabel = "This one is labelled", lw = 3, title = "Subtitle")
p4 = histogram(x, y) # Four histograms each with 10 points? Why not!
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

```julia
using StatsPlots # Required for the DataFrame user recipe
# Now let's create the DataFrame
using DataFrames
df = DataFrame(a = 1:10, b = 10 * rand(10), c = 10 * rand(10))
# Plot the DataFrame by declaring the points by the column names
@df df plot(:a, [:b :c]) # x = :a, y = [:b :c]. Notice this is two columns!
```

```julia
@df df scatter(:a, :b, title = "My DataFrame Scatter Plot!") # x = :a, y = :b
```

```julia
using Distributions
plot(Normal(3, 5), lw = 3)
```

```julia
using RDatasets, StatsPlots
iris = dataset("datasets", "iris")
@df iris marginalhist(:PetalLength, :PetalWidth)
```

```julia
y = rand(100, 4) # Four series of 100 points each
violin(["Series 1" "Series 2" "Series 3" "Series 4"], y, leg = false)
```

```julia
boxplot!(["Series 1" "Series 2" "Series 3" "Series 4"], y, leg = false)
```
